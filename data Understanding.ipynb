# Isi per Trace

```python
# Initialize dictionaries to store counts
network_counts = {}
channel_counts = {}
location_counts = {}
station_counts = {}

# Iterate over each trace in the stream
for trace in stream:
    network = trace.stats.network
    channel = trace.stats.channel
    location = trace.stats.location
    station = trace.stats.station

    # Update network counts
    if network in network_counts:
        network_counts[network] += 1
    else:
        network_counts[network] = 1

    # Update channel counts
    if channel in channel_counts:
        channel_counts[channel] += 1
    else:
        channel_counts[channel] = 1

    # Update location counts
    if location in location_counts:
        location_counts[location] += 1
    else:
        location_counts[location] = 1
        
    # Update station counts
    if station in station_counts:
        station_counts[station] += 1
    else:
        station_counts[station] = 1

# Print the counts
print("Network counts:")
for network, count in network_counts.items():
    print(f"{network} = {count}")

print("\nChannel counts:")
for channel, count in channel_counts.items():
    print(f"{channel} = {count}")

print("\nLocation counts:")
for location, count in location_counts.items():
    print(f"{location} = {count}")
    
print("\nStation counts:")
for station, count in station_counts.items():
    print(f"{station} = {count}")

print(f"\nTraces counts: {len(stream)}")
```

# Persebaran Metadata per Folder

```python
from obspy import read
import os

# Define the folder containing MSEED files
folder_path = "D:\Skripsi\SKRIPSI-FINAL-CODE\Dataset\Dataset Modeling\BOYONG"  #YG ORI

# Initialize dictionaries to store counts
network_counts = {}
channel_counts = {}
location_counts = {}
station_counts = {}

# Iterate over each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".msd"):
        file_path = os.path.join(folder_path, filename)
        # Read the MSEED file
        stream = read(file_path)
        
        # Iterate over each trace in the stream
        for trace in stream:
            network = trace.stats.network
            channel = trace.stats.channel
            location = trace.stats.location
            station = trace.stats.station

            # Update network counts
            if network in network_counts:
                network_counts[network] += 1
            else:
                network_counts[network] = 1

            # Update channel counts
            if channel in channel_counts:
                channel_counts[channel] += 1
            else:
                channel_counts[channel] = 1

            # Update location counts
            if location in location_counts:
                location_counts[location] += 1
            else:
                location_counts[location] = 1

            # Update station counts
            if station in station_counts:
                station_counts[station] += 1
            else:
                station_counts[station] = 1

# Print the counts
print("Network counts:")
for network, count in network_counts.items():
    print(f"{network} = {count}")

print("\nChannel counts:")
for channel, count in channel_counts.items():
    print(f"{channel} = {count}")

print("\nLocation counts:")
for location, count in location_counts.items():
    print(f"{location} = {count}")

print("\nStation counts:")
for station, count in station_counts.items():
    print(f"{station} = {count}")
```
